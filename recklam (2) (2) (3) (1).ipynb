```python
#The Central Federal District
row = travel.iloc[0]
row_df = pd.DataFrame([row])
ts_travel_3 = row_df.melt(var_name='Month', value_name='Trips')

ts_travel_3['Month'] = pd.Categorical(ts_travel_3['Month'])
ts_travel_3['Month_code'] = ts_travel_3['Month'].cat.codes
X3 = ts_travel_3['Month_code'].values.reshape(-1, 1)
y3 = ts_travel_3['Trips'].values
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
model = CatBoostRegressor(loss_function='RMSE')
model.fit(X3_train, y3_train, verbose=False)

score = model.score(X3_test, y3_test)
print(f"R-squared score: {score:.2f}")

model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.1,
    depth=6,
    reg_lambda=3
)
model.fit(X3_train, y3_train, verbose=False)
new_month = 8  # Июнь
y_pred = model.predict([[new_month]])
print(f"Прогнозируемое количество поездок в {ts_travel_3['Month'].cat.categories[new_month]}: {int(y_pred[0])}")
```
